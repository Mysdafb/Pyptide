### Setters

To change the values of the attributes, the class Preprocessor provides setter functions:

1. set_names(list[names]): assigns an ID to each sequence. Its argument is a python list with names for each sequence; the length of the list of names should be the same for the sequence list.
2. set_descriptors([list[descriptors], str[filename:csv]], bool[headers]): assigns a molecular descriptor or a set of molecular descriptors to each sequence. Its argument can be a python list containing the molecular descriptors for each sequence or a CSV file with this information. The 'headers' argument should be True if the file contains column names and false if not.
3. set_features_names(list[features]): stores a list of names for the molecular descriptors associated with each sequence. Its argument is a Python list containing the names of the features.
4. set_targets([list[targets], str[filename:txt,cvs]], bool[headers]): assigns a label (class) to each sequence. Its argument can be a python list with each label for each sequence or a file with CSV or txt extension containing the labels in a single column. The 'headers' argument should be True if the file contains column names and false if not.

```python
Seqs.set_names(['>'+str(i) for i in range(len(Seqs.get_sequences()))])
Seqs.get_names()
```

```python
Seqs.set_descriptors('./descriptors.csv')
Seqs.get_descriptors()
```

```python
Seqs.set_features_names(['PAAC'+str(i) for i in range(1,23)])
Seqs.get_feature_names()
```

#### Filtering

The class Preprocessor provides some methods to perform data cleaning. Those methods allow us to filter sequences by length and filter by unnatural amino acids.

1. filter_by_length(int[min],int[max]): eliminates sequences from loaded data that are not in the range [min, max]. Its arguments are two integers numbers expressing the min and max amino acid length of sequences. By default min=5 and max=100.
2. filter_unnatural_aa(): eliminates sequences that contain unnatural amino acids. Optionally, you can pass the argument return_ = True; to get in a python list the sequences dropped.

Additionally, is given the clean() method, it applies the two methods above, at the same time, with its default arguments.

```python
Seqs.filter_by_length(10, 20)
Seqs.get_number_of_seqs()
```

```python
Seqs = Preprocessor('./anticancer_peptides.fasta')
unnaturals = Seqs.filter_unnatural_aa(True)
print(unnaturals)
```

```python
Seqs = Preprocessor('./anticancer_peptides.fasta')
Seqs.clean()
Seqs.get_number_of_seqs()
```

The class Preprocessor also provides a method to eliminate duplicated sequences.

1. filter_duplicates(): remove duplicate sequences, leaving only one copy of them.

```python
Seqs.filter_duplicates()
Seqs.get_number_of_seqs()
```

#### Building a negative dataset

The class Preprocessor has a method to generate a negative dataset with the same length distribution of the positive sequences (for binary classification), following the methodology described in Gabere and Noble (2017).

1. get_decoys(str[filename:fasta],int[ratio],int[seed]): stores in your local machine a fasta file containing decoys samples. Its arguments are a fasta file with the sequences from what decoys will be generated; a ratio of positive:negative sequence to generate and a random seed to reproducibility.

The name of the fasta file stored should be the same name passing as the first argument plus the '_decoys.fasta' suffix, and you can find this file in the same directory.

```python
Seqs = Preprocessor('./anticancer_only_pos.fasta') #Loading a positive dataset
Seqs.get_decoys('./uniprot.fasta', ratio=1, seed=7000) #uniprot.fasta contains proteins from uniprot database as describe in Gabere and Noble (2017)  
```

#### Identical overlap with other datasets

To identify whether your dataset has an identical overlap (intersection) with another dataset, you can use the following method:

1. identical_overlap(str[filename:fasta], bool[remove_from_seqs]): finds the identical overlap between the sequences loaded and the file given as argument. Its argument is a path to the fasta file to compare, and True, if you want to remove from the Preprocessor object the sequences in both datasets, False in other case. If false, return a python list with the sequences in both datasets

```python
Seqs = Preprocessor('./anticancer_peptides.fasta')
Seqs.identical_overlap('./uniprot_decoys.fasta', remove_from_seqs=True)#Check the overlap between positive and negative sets
Seqs.get_number_of_seqs() #Check overlap
```

#### Selection and splitting

To select a subset of sequences is provided the functions below:

1. select_seqs_at_rand(int[n], bool[only_return]): selects at random n sequences from the Preprocessor object. Its arguments are an integer number, corresponding to the quantity of sequence to select, and True if you want to remove unselected sequences from the object, or False if only want to obtain a python list with chosen sequences.
2. split_train_test_random(float[p], bool[only_return]): divides the dataset into two subsets according to the proportion given. Its arguments are a float number corresponding to the proportion of the training subset (the remaining sequences are assigned to the testing set), and True if you want to leave only the training subset in the Preprocessor object and return a python list with the testing subset, False in another case, this option return both training and testing subsets in a python list.

### Molecular descriptors calculation

The class Preprocessor also includes a set of methods to calculate molecular descriptors to the sequences loaded. 

1. aminoacid_composition(): computes the amino acid composition proportion of each sequence.
2. aliphatic_index(): calculates the aliphatic index for each sequence.
3. aromaticity(): computes the aromaticity index for each sequence.
4. calculate_charge(): computes the charge of each sequence.
5. composition(): computes the composition of different groups of amino acids of each sequence.
6. distribution(): computes the distribution of different groups of amino acids of each sequence.
7. hydrophobic_ratio(): computes the hydrophobic ratio of each sequence.
8. isoelectric_point(): computes the isoelectric point of each sequence.
9. molecular_weight(): computes the molecular weight of each sequence.
10. pseudo_aminoacid_composition(int[lambda_value], float[w]): computes the pseudo amino acid composition of each sequence. By default lambda_value=2 and w=0.05.
11. transition(): computes the transition of different groups of amino acids of each sequence.
12. calculate_all(): computes all the above molecular descriptors; in total, 321 molecular descriptors are calculated.

All methods have the argument 'append'; if True, then the molecular descriptors add to the previous one calculated. If False, the previous one is replaced. For more information about the methods provided, see the python packages modlamp and ilearn, from which are based.

```python
Seqs = Preprocessor('./anticancer_peptides.fasta') #Loading a dataset
Seqs.calculate_all()
```

All methods have the argument 'save'; if True, the plot is saved in your working directory.

### Pre-modeling

To get data ready to modeling, the class Preprocessor has the method below:

1. scaling(str[scaler]): Scales the molecular descriptors based on the algorithm selected, and save the parameters, used to scale, in the attributes numerator and denominator of the class Preprocessor. Its argument is a string represented the algorithm to choose. The types of algorithms are:

    mean-std:    (x - mean) / std
    
    range:       x / range;  where range = max - min
    
    mean-range   (x - mean) / range; where range = max - min
    
    min-max      (x - min) / max
    
By default the algorithm is mean-std.

```python
Seqs.calculate_all() #Calculate all descriptors available
Seqs.get_descriptors()[:5]#molecular descriptors before scaling
```

```python
Seqs.scaling('mean-range')
Seqs.descriptors[:5]#molecular descriptors after scaling
```

### Separability

To analyze the separability of the datasets, the class Preprocessor provides the following method:

1. sequence_based_clustering(): Performs an Espectral Clustering based on Levenshtein distance (sequence-based similarity), the number of clusters is given by the number of labels in Preprocessor.targets property. This method, return validation metrics in the following order: sensitivity, specificity, accuracy, Matthews correlation coefficient, AUC-ROC, and the kappa statistic.

```python
sn, sp, acc, mcc, aucroc, k = Seqs.sequence_based_clustering()
print(sn, sp, acc, mcc, aucroc, k)
```

### Modeling

To select and built a classification model, the class Preprocessor provides the following functions:

1. find_best_model(str[algorithm], dict[search_space], int[cv], str[scaler]): performs a grid search algorithm over the search_space defined, evaluates each model in a k-folds cross validation process, and stores the best model in the attribute model of the class Preprocessor. If the scaler parameter is given, before the grid search runs, the molecular descriptors are scaled.
2. train(): train a clasification model stored in the model attribute using all the data.

```python
Seqs.pseudo_aminoacid_composition() #Calculate PseAAC molecular descriptors
Seqs.find_best_model('rf') #available models: rf, svm, knn
```

```python
Seqs.train() #train the model selected above with all the data (descriptors) in the class Preprocessor
```

### Evaluation

To evaluate the model selected, the class Preprocessor provides the following functions:

1. k_fold_cross_validation(int[folds], str[scaler], bool[history]): Performs a k-folds cross-validation procedure (with k=folds) and returns the average metrics obtained in this process. If is given an algorithm to scale, before runs the k-fold cross-validation, the data are scaled. If history = True, return a list of lists with the metrics gotten in each fold (by default history = False). The metrics are returned in the next order: sn, sp, acc, mcc, auc-roc, k.
2. repeated_holdout_validation(int[nrepetitions], float[p], str[scaler], bool[history]): Performs a repeated hold-out procedure and returns the average metrics obtained in this process. The argument p gives the size of the hold-out set. If is given an algorithm to scale, before runs the k-fold cross-validation, the data are scaled. If history = True, return a list of lists with the metrics gotten in each fold (by default history = False). The metrics are returned in the next order: sn, sp, acc, mcc, auc-roc, k.
3. random_feature_validation(): To evaluate spuriuos correlation of the molecular descriptors with the targets, this method generates for each sequence a random value of each molecular descriptor between the range (max-min) of this, then performs a 10-fold cross-validation. Returns the average metrics obtained in the 10-fold cross-validation.
4. y_randomization_validation(float[p]): scramble the p proportion of the predicted variable.


```python
Seqs.k_fold_cross_validation(10)
```

```python
Seqs.repeated_holdout_validation(10, 0.2)
```

```python
Seqs.random_feature_validation()
```

```python
Seqs.y_randomization_validation(0.4) #By default the argument p=0.1
```
